# Age group

```python
print('Median: ',facebook_data['age'].median())
print('Mean: ',facebook_data['age'].mean())
```

```python
plt.figure(figsize=(8,6))
sns.distplot(facebook_data['age'], kde=False)
```

# Likes Received

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="age",y="likes_received",hue="gender",data=facebook_data)
```

We can conclude that the females aged below 60 receive much more amount of likes than males, after than the amount of likes they receive are similar. The females of age group around 20 receive significantly more amount of likes than male. 

# Gender with more account

```python
print('Male: ',facebook_data[facebook_data['gender']=='male'].shape[0])

print('Female: ',facebook_data[facebook_data['gender']=='female'].shape[0])
```

Here, the number of male using facebook is higher than female, so we can conclude a larger fraction of demographic of facebook is male.

# Gender and friends

```python
sns.barplot(x='gender',y ='friend_count', data=facebook_data)
```

```python
print("Average Friends of Male: ", int(facebook_data[facebook_data["gender"]=='male']['friend_count'].mean()))
print("Average Friends of Female: ", int(facebook_data[facebook_data["gender"]=='female']['friend_count'].mean()))
```

From this, we can conclude that female have more number of friends than male in general.

# Mobile Likes vs Web Likes

```python
mobile_likes=facebook_data["mobile_likes"].sum()
```

```python
web_likes=facebook_data["www_likes"].sum()
```

```python

web_mobile = sns.barplot(x=["mobile","web"],y=[mobile_likes,web_likes])
web_mobile.text(0,mobile_likes/2,str(round(mobile_likes*100/(mobile_likes+web_likes),2))+'%',fontdict={'color':"white", 'fontsize':15},horizontalalignment='center')
web_mobile.text(1,web_likes/2,str(round(100-mobile_likes*100/(mobile_likes+web_likes),2))+'%',fontdict={'color':"white", 'fontsize':15},horizontalalignment='center')
```

We can see that 67.99% are mobile likes whereas only 32.01% are web likes. Thus, we can conclude that more users use fb on mobile than on web.
